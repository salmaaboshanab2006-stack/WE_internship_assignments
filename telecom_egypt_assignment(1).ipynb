## Question 5 — Analyze Data Usage

Write a **function** called `analyze_usage` that takes a list of data usage values (in GB)  
and returns a dictionary containing:
- `average` — the average usage
- `max` — the highest usage
- `min` — the lowest usage

Then call the function with this list and print the results:
```python
usage = [4.5, 12.0, 7.3, 9.8, 2.1, 15.0, 6.6]
```

**Expected Output Example:**
```
Average Usage : 8.19 GB
Max Usage     : 15.0 GB
Min Usage     : 2.1 GB
```



````python
# Question 5 — Write your solution here
def analyze_usage(usage):
    result = {
        "average": sum(usage) / len(usage),
        "max": max(usage),
        "min": min(usage)
    }
    return result


usage = [4.5, 12.0, 7.3, 9.8, 2.1, 15.0, 6.6]

result = analyze_usage(usage)

print("Average Usage :", round(result["average"], 2), "GB")
print("Max Usage     :", result["max"], "GB")
print("Min Usage     :", result["min"], "GB")
````

## Question 6 — Save Complaints to a File

Write a program with two functions:

1. `add_complaint(phone, complaint)` — saves the complaint to a file called `complaints.txt`  
   in the format: `phone | complaint`

2. `show_complaints()` — reads and prints all saved complaints from the file.  
   If the file does not exist, print: `No complaints found.`

Ask the user for their phone number and complaint, then call both functions.

**Expected Output Example:**
```
Complaint saved successfully!

--- All Complaints ---
01512111111 | Internet is very slow
01501000567 | Cannot make calls
```


````python
# Question 6 — Write your solution here
def add_complaint(phone, complaint):
    file = open("complaints.txt", "a")
    file.write(phone + " | " + complaint + "\n")
    file.close()
    print("Complaint saved successfully!")


def show_complaints():
    try:
        file = open("complaints.txt", "r")
        print("\n--- All Complaints ---")
        print(file.read())
        file.close()
    except FileNotFoundError:
        print("No complaints found.")


phone = input("Enter phone number: ")
complaint = input("Enter complaint: ")

add_complaint(phone, complaint)
show_complaints()
````

## Question 7 — Subscriber Management System (OOP)

### Part A — Base Class

Create a class called `Subscriber` with:

**Attributes (`__init__`):**
- `name`, `phone`, `package` (Bronze / Silver / Gold), `balance` (default = 0)

**Package Prices (store inside the class):**
| Package | Monthly Price |
|---------|--------------|
| Bronze  | 50 EGP       |
| Silver  | 100 EGP      |
| Gold    | 200 EGP      |

**Methods:**
- `recharge(amount)` — adds amount to balance, prints new balance.  
  Raise `ValueError` if amount ≤ 0.
- `deduct(amount)` — deducts amount from balance.  
  Raise `ValueError` if amount ≤ 0.  
  Raise `Exception` if balance is insufficient.
- `upgrade_package(new_package)` — changes the package **only if**:
  1. `new_package` is valid (Bronze / Silver / Gold)
  2. Balance is enough to cover the new package price  
  Deduct the price and confirm. If not enough → print how much is needed.
- `show_info()` — prints all subscriber details including balance.
- `__str__()` — returns a formatted string of the subscriber (used when printing the object directly).

---

### Part B

Create two subclasses:

**`PrepaidSubscriber(Subscriber)`:**
- Has an extra attribute: `data_gb` (default = 0)
- Override `recharge(amount)`: after recharging, add 1 GB for every 10 EGP recharged.
- Add method `check_data()`: prints remaining data in GB.

**`PostpaidSubscriber(Subscriber)`:**
- Has an extra attribute: `credit_limit` (default = 500 EGP)
- Override `deduct(amount)`: allow balance to go **negative** but not below `-credit_limit`.  
  If it exceeds the limit → raise `Exception("Credit limit reached!")`.
- Add method `show_debt()`: if balance is negative, print how much the subscriber owes.

---

### Part C

1. Create a list called `subscribers` containing:
   - 2 `PrepaidSubscriber` objects
   - 2 `PostpaidSubscriber` objects
   - 1 `Subscriber` object (base)

3. Write a function `find_by_phone(subscribers, phone)`:
   - Returns the subscriber object if found.
   - Raises `ValueError("Subscriber not found.")` if not.

4. Write a function `filter_by_package(subscribers, package)`:
   - Returns a list of all subscribers on that package.

5. Write a function `top_balance(subscribers)`:
   - Returns the subscriber with the highest balance.

### Part D

Using the list from Part C:
1. Find a subscriber by phone → upgrade their package.
2. Recharge a `PrepaidSubscriber` → check their data.
3. Deduct from a `PostpaidSubscriber` until credit limit is hit → handle the exception.
4. Filter all `Gold` subscribers and print them.
5. Print the subscriber with the top balance.

```
**Expected Output Example:**

[Postpaid] Karim Sami  | 01525556789 | Gold   | Balance: 0.00 EGP  | Limit: 500 EGP
```

````python
# ==========================
# PART A
# ==========================

class Subscriber:
    package_prices = {
        "Bronze": 50,
        "Silver": 100,
        "Gold": 200
    }

    def __init__(self, name, phone, package, balance=0):
        self.name = name
        self.phone = phone
        self.package = package
        self.balance = balance

    def recharge(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be greater than 0.")
        self.balance += amount
        print("New Balance:", self.balance)

    def deduct(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be greater than 0.")
        if self.balance < amount:
            raise Exception("Insufficient balance.")
        self.balance -= amount

    def upgrade_package(self, new_package):
        if new_package not in Subscriber.package_prices:
            print("Invalid package.")
            return

        price = Subscriber.package_prices[new_package]

        if self.balance >= price:
            self.deduct(price)
            self.package = new_package
            print("Package upgraded successfully!")
        else:
            print("Need", price - self.balance, "EGP more.")

    def show_info(self):
        print("Name:", self.name)
        print("Phone:", self.phone)
        print("Package:", self.package)
        print("Balance:", self.balance)

    def __str__(self):
        return f"Name: {self.name}, Phone: {self.phone}, Package: {self.package}, Balance: {self.balance} EGP"


# ==========================
# PART B
# ==========================

class PrepaidSubscriber(Subscriber):

    def __init__(self, name, phone, package, balance=0, data_gb=0):
        super().__init__(name, phone, package, balance)
        self.data_gb = data_gb

    def recharge(self, amount):
        super().recharge(amount)
        self.data_gb += amount // 10
        print("Data:", self.data_gb, "GB")

    def check_data(self):
        print("Remaining Data:", self.data_gb, "GB")


class PostpaidSubscriber(Subscriber):

    def __init__(self, name, phone, package, balance=0, credit_limit=500):
        super().__init__(name, phone, package, balance)
        self.credit_limit = credit_limit

    def deduct(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be greater than 0.")

        if self.balance - amount < -self.credit_limit:
            raise Exception("Credit limit reached!")

        self.balance -= amount

    def show_debt(self):
        if self.balance < 0:
            print("Debt:", -self.balance, "EGP")
        else:
            print("No debt.")
    # ==========================
# PART C
# ==========================

subscribers = [
    PrepaidSubscriber("Salma", "01011111111", "Bronze", 100, 5),
    PrepaidSubscriber("Ali", "01022222222", "Silver", 200, 10),
    PostpaidSubscriber("Mona", "01033333333", "Gold", 50),
    PostpaidSubscriber("Omar", "01044444444", "Bronze", -100),
    Subscriber("Sara", "01055555555", "Silver", 150)
]

print("----- All Subscribers -----")
for s in subscribers:
    print(s)


def find_by_phone(subscribers, phone):
    for s in subscribers:
        if s.phone == phone:
            return s
    raise ValueError("Subscriber not found.")


def filter_by_package(subscribers, package):
    result = []
    for s in subscribers:
        if s.package == package:
            result.append(s)
    return result


def top_balance(subscribers):
    top = subscribers[0]
    for s in subscribers:
        if s.balance > top.balance:
            top = s
    return top


# ==========================
# PART D
# ==========================

print("\n1. Find Subscriber By Phone")
try:
    s = find_by_phone(subscribers, "01011111111")
    s.upgrade_package("Silver")
    s.show_info()
except ValueError as e:
    print(e)

print("\n2. Recharge Prepaid Subscriber")
subscribers[0].recharge(50)
subscribers[0].check_data()

print("\n3. Postpaid Deduction")
try:
    while True:
        subscribers[2].deduct(200)
        print("Balance:", subscribers[2].balance)
except Exception as e:
    print(e)

subscribers[2].show_debt()

print("\n4. Gold Subscribers")
gold = filter_by_package(subscribers, "Gold")
for s in gold:
    print(s)

print("\n5. Subscriber With Highest Balance")
print(top_balance(subscribers))
````
